# Build the strategy

```python
class testSTG(bt.Strategy):
     def __init__(self):
        pass
     def next(self):
        pass
    
class testClose(bt.Strategy):
    def __init__(self):
        pass
    def next(self):
        dataclose=self.datas[0].close[0]       #internal referencing slide
        print(dataclose)
        
class testClosei(bt.Strategy):
    def __init__(self):
        self.dataclose=self.datas[0].close    # took dataclose into the initilizer  #init= initilizaer
    def next(self):
        print(self.dataclose[0])
        
       
class testLog(bt.Strategy):     #to notify our trader about trade
    def __init__(self):
        self.dataclose=self.datas[0].close
    def log(self,txt):
        dt=self.datas[0].datetime.datetime()
        print(f'{dt}|{txt}')
    def next(self):
        x= self.dataclose[0]
        self.log(txt=x)
```

```python
cerebro=bt.Cerebro()
```

```python
cerebro.adddata(data)
```

```python
cerebro.addstrategy(testSTG)
```

```python
cerebro.run()
```
